```python
# plot a 2d kde estimate of a set of points
def kde2d(x,y,cont=True,points=False):
    xmin,xmax = min(x),max(x)
    ymin,ymax = min(y),max(y)
    gx,gy = mgrid[xmin:xmax:100j,ymin:ymax:100j]
    positions = vstack([gx.ravel(),gy.ravel()])
    values = vstack([x,y])
    kernel = stats.gaussian_kde(values)
    f = reshape(kernel(positions).T,gx.shape)
    # plot
    xlim(xmin,xmax)
    ylim(ymin,ymax)
    imshow(np.rot90(f), cmap=plt.cm.gist_earth_r, extent=[xmin, xmax, ymin, ymax])
    if contour:
        cset = contour(gx,gy,f)
        #clabel(cset, inline=1, fontsize=10)
        colorbar()
    if points:
        scatter(x, y, c='r', s=35)

    
    
    
```

```python
# Generative model. The latent variable z acts as a switch that determines which gaussian each data point is drawn from
from scipy import stats
samples = 10000
pz = stats.rv_discrete(name="pz",values=((0,1),(0.3,0.7))) # creates a discrete distribution I can sample from
u = [[0,0],[5,5]] # means
sigmas = [array([[1,.3],[0,.1]]),array([[1,0],[.4,1]])]
X = zeros((samples,2))
for i in xrange(samples):
    z = pz.rvs() # sample a value from P(z)
    x_i = multivariate_normal(u[z],sigmas[z]) # sample x from the corresponding Gaussian
    X[i] = x_i

hist2d(X[:,0],X[:,1],bins=40)
show()

```

```python
# Lets try Gradient Descent, We will let W stand for U and H stand for Vt. So we are trying to find W,H s.t M = WH

def error(M,W,H):
    E = M - dot(W,H)
    return sum(pow(E,2)) # this is the Frobius norm squared

def alternate_update(M,W,H,W_update,H_update,iterations):gradient_solve
    errors = []
    for i in xrange(iterations):
        errors.append(error(M,W,H))
        H2 = H_update(M,W,H)
        W2 = W_update(M,W,H)
        H = H2
        W = W2
        
    xlabel("iteration")
    ylabel("error")
    title("Convergence of Gradient Decsent")
    plot(range(iterations),errors)
    show()
    M2 = dot(W,H) # final estimate of M
    
    minW,minH = np.min(W),np.min(H)
    print "Non-negative",minW > 0 and minH > 0, "Smallest W",minW,"Smallest H",minH
    print "Final reconstruction error",errors[-1]
    return (M2,W,H)

def plot_factorization(M,M2,U,W):
    # Two subplots, the axes array is 1-d
    f, axarr = plt.subplots(2, sharex=True)
    axarr[0].plot(M)
    axarr[0].set_title('Original & recononstructed data')
    axarr[1].plot(M2)
    show()

    f, axarr = plt.subplots(2, sharex=True)
    axarr[0].plot(U)
    axarr[0].set_title('Original & recononstructed basis vectors')
    axarr[1].plot(W)
    show()


# additive updates ---------------
def updateH(M,W,H,gamma = 0.001):
    Wt = transpose(W)
    a = dot(Wt,M)
    b = dot(Wt,dot(W,H))
    H2 = H +  gamma*(a - b)
    return H2

def updateW(M,W,H,gamma = 0.001):
    Ht = transpose(H)
    a = dot(M,Ht)
    b = dot(W,dot(H,Ht))
    W2 = W + gamma*(a - b)
    return W2    
# ---------------------------------

# multiplicative updates ----------
def mUpdateH(M,W,H):
    Wt = transpose(W)
    a = dot(Wt,M)
    b = dot(Wt,dot(W,H))
    H2 = H*(a/b)
    return H2
    
def mUpdateW(M,W,H):
    Ht = transpose(H)
    a = dot(M,Ht)
    b = dot(W,dot(H,Ht))
    W2 = W*(a/b)
    return W2
# ---------------------------------
    
# initialize with random positive values
W = np.random.random(size=(m,k))
H = np.random.random(size=(k,n))

M2,W2,H2 = alternate_update(M,W,H,updateW,updateH,10000)
plot_factorization(M,M2,U,W2)

M3,W3,H3 = alternate_update(M,W,H,mUpdateW,mUpdateH,100)
plot_factorization(M,M3,U,W3)
```
